```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure numeric columns
shipping['Orders'] = pd.to_numeric(shipping['Orders'], errors='coerce')
shipping['Revenue'] = pd.to_numeric(shipping['Revenue'], errors='coerce')
shipping['Orders (previous_period)'] = pd.to_numeric(shipping['Orders (previous_period)'], errors='coerce')
shipping['Total sales (previous_period)'] = pd.to_numeric(shipping['Total sales (previous_period)'], errors='coerce')

# -------------------------
# KPIs Calculation
# -------------------------
total_orders = shipping['Orders'].sum()
total_revenue = shipping['Revenue'].sum()
total_orders_prev = shipping['Orders (previous_period)'].sum()
total_revenue_prev = shipping['Total sales (previous_period)'].sum()

orders_growth = ((total_orders - total_orders_prev) / total_orders_prev) * 100
revenue_growth = ((total_revenue - total_revenue_prev) / total_revenue_prev) * 100

# Revenue per Order by Channel
revenue_per_order = shipping.groupby('Referring Channel').apply(
    lambda x: x['Revenue'].sum() / x['Orders'].sum()
).sort_values(ascending=False)

# Top Channels by Revenue
top_channels = shipping.groupby('Referring Channel')['Revenue'].sum().sort_values(ascending=False)

# -------------------------
# Print KPIs
# -------------------------
print("=== Shipping / Marketing Channel KPIs ===")
print(f"Total Orders: {total_orders}")
print(f"Total Revenue: ${total_revenue:,.2f}")
print(f"Orders Growth vs Previous Period: {orders_growth:.2f}%")
print(f"Revenue Growth vs Previous Period: {revenue_growth:.2f}%\n")

print("Revenue per Order by Channel:")
print(revenue_per_order)
print("\nTop 5 Channels by Revenue:")
print(top_channels.head(5))

# -------------------------
# Visualization
# -------------------------
sns.set(style="whitegrid")

# 1. Top 5 Channels by Revenue
plt.figure(figsize=(10,6))
sns.barplot(x=top_channels.head(5).values, y=top_channels.head(5).index, palette="Blues_r")
plt.title("Top 5 Marketing Channels by Revenue")
plt.xlabel("Revenue")
plt.ylabel("Channel")
plt.show()

# 2. Revenue per Order by Channel
plt.figure(figsize=(10,6))
sns.barplot(x=revenue_per_order.values, y=revenue_per_order.index, palette="Greens_r")
plt.title("Revenue per Order by Channel")
plt.xlabel("Revenue per Order")
plt.ylabel("Channel")
plt.show()

# 3. Growth Comparison (Orders vs Revenue)
growth_df = pd.DataFrame({
    "Metric": ["Orders Growth", "Revenue Growth"],
    "Growth (%)": [orders_growth, revenue_growth]
})

plt.figure(figsize=(6,4))
sns.barplot(x="Metric", y="Growth (%)", data=growth_df, palette="Oranges_r")
plt.title("Orders and Revenue Growth vs Previous Period")
plt.ylim(min(0, growth_df['Growth (%)'].min()-5), growth_df['Growth (%)'].max()+5)
plt.show()
```
